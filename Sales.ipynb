```python
df = df.drop(columns=['Price'])
```

```python
def Pricing(x):
    if x == 'Maxi MTA':
        return 80
    elif x == 'Sunset':
        return 27
    elif x == 'Yanaki':
        return 25
    elif x == 'Sunbell':
        return 24
    elif x == 'Aspen':
        return 25
    elif x == 'Sunshine':
        return 26
    elif x == 'Quad':
        return 43
    elif x == 'TriFly':
        return 4
    elif x == 'Carlota':
        return 27
    elif x == 'Doublers':
        return 125
    elif x == 'Distance':
        return 75
    elif x == 'FastCatch':
        return 50
```

```python
df['Price'] = df['Product'].apply(Pricing)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Revenue'] = df['Units'] * df['Price'] * (1 - df['Discount'])
```

```python
df.info()
```

```python
def Costs(x):
    if x == 'Maxi MTA':
        return 35
    elif x == 'Sunset':
        return 10
    elif x == 'Yanaki':
        return 11
    elif x == 'Sunbell':
        return 9
    elif x == 'Aspen':
        return 11
    elif x == 'Sunshine':
        return 10
    elif x == 'Quad':
        return 19
    elif x == 'TriFly':
        return 2
    elif x == 'Carlota':
        return 12
    elif x == 'Doublers':
        return 55
    elif x == 'Distance':
        return 33
    elif x == 'FastCatch':
        return 21
```

```python
df['COGS'] = df['Product'].apply(Costs)
```

```python
df['Net_Rev'] = df['Revenue'] - (df['COGS'] * df['Units'])
```
